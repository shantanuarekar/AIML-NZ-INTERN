Task 1: Python Basics & Operators
1. Create a Python script that accepts two numbers from the user.
2. Perform all basic arithmetic operations (add, subtract, multiply, divide, modulus, power).
3. Demonstrate use of comparison and logical operators (>, <, ==, !=, and, or, not).
4. Print results clearly using formatted output (f-strings).

```python
import random
import numpy as np
from datetime import datetime

# Task 1: Python Basics & Operators
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print(f"Addition: {num1 + num2}")
print(f"Subtraction: {num1 - num2}")
print(f"Multiplication: {num1 * num2}")
print(f"Division: {num1 / num2}")
print(f"Modulus: {num1 % num2}")
print(f"Power: {num1 ** num2}")

print(f"{num1} > {num2}: {num1 > num2}")
print(f"{num1} < {num2}: {num1 < num2}")
print(f"{num1} == {num2}: {num1 == num2}")
print(f"{num1} != {num2}: {num1 != num2}")
```

Task 2: Lists and Arrays
1. Create a list of 10 random integers.
2. Perform the following operations:
○ Add and remove elements.
○ Find the maximum and minimum values.
○ Sort the list.
3. Convert the list into a NumPy array and calculate:
○ Mean, Median, and Standard Deviation.

```python
# Task 2: Lists and Arrays
import random
import numpy as np
from datetime import datetime
numbers = [random.randint(1, 100) for _ in range(10)]
numbers.append(50)
numbers.pop(0)

print("List:", numbers)
print("Max:", max(numbers))
print("Min:", min(numbers))

array = np.array(numbers)
print("Mean:", np.mean(array))
print("Median:", np.median(array))
print("Std Dev:", np.std(array))
```

```python
# Task 3: Dictionary & Sets
import random
import numpy as np
from datetime import datetime
student = {"name": "Shaantanu", "course": "AI/ML", "marks": 82}

if student["marks"] >= 80:
    student["grade"] = "A"
elif student["marks"] >= 60:
    student["grade"] = "B"
else:
    student["grade"] = "C"

for k, v in student.items():
    print(k, ":", v)

ai_tools_1 = {"ChatGPT", "TensorFlow", "PyTorch"}
ai_tools_2 = {"PyTorch", "Keras", "Scikit-learn"}
print("Common AI Tools:", ai_tools_1 & ai_tools_2)
```

Task 5: Real-World Mini Project
Create a Python program named ai_prompt_logger.py that:
1. Accepts a user input (simulating an AI prompt).
2. Saves the prompt text in a file named prompt_history.txt.
3. Each time you run the program, new prompts are appended with timestamps.

```python
# Task 5: Prompt Logger
prompt = input("Enter AI prompt: ")
timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

with open("prompt_history.txt", "a") as f:
    f.write(f"[{timestamp}] {prompt}\n")

print("Prompt saved!")
```
