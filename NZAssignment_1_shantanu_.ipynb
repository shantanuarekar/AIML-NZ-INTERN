```python
import nltk
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')
```

```python
para1 = """The rapid growth of digital communication has transformed the way people interact and share information across the world. Social media platforms, emails, and messaging applications generate massive amounts of textual data every day. Analyzing this data can provide valuable insights into user behavior, opinions, and trends. However, raw text often contains noise such as punctuation, stop words, and inconsistent formatting, which makes preprocessing an essential step before applying any Natural Language Processing techniques."""

para2 = """
In addition to communication data, textual information from news articles, blogs, and customer reviews plays a crucial role in decision-making processes. Businesses rely on this data to understand customer sentiment, improve products, and enhance user experience. Techniques like tokenization, stemming, lemmatization, and word embeddings help convert unstructured text into meaningful representations. By applying these methods, machines can identify patterns, extract key information, and support tasks such as classification, clustering, and recommendation systems."""

text = para1 + " " + para2

print("Original Text:\n")
print(text)
```

## **Removing punctuation marks**





```python
text_no_punct = text.translate(str.maketrans('', '', string.punctuation))

print("Text after removing punctuation:\n")
print(text_no_punct)
```

# **Stop word removal**

```python
stop_words = set(stopwords.words('english'))

words = text_no_punct.split()

filtered_words = []

for word in words:
    if word.lower() not in stop_words:
        filtered_words.append(word)

print("After removing stop words:\n")
print(filtered_words)
```

# **Stemming**

```python
stemmer = PorterStemmer()

stemmed_words = []

for word in filtered_words:
    stemmed_words.append(stemmer.stem(word))

print("After Stemming:\n")
print(stemmed_words)
```

# **Lemmatization**

```python
lemmatizer = WordNetLemmatizer()

lemmatized_words = []

for word in filtered_words:
    lemmatized_words.append(lemmatizer.lemmatize(word))

print("After Lemmatization:\n")
print(lemmatized_words)
```

**ASSIGNMENT-2**  Feature Engineering

```python
from sklearn.feature_extraction.text import CountVectorizer

text = ["""The rapid growth of digital communication has transformed the way people interact and share information across the world. Social media platforms, emails, and messaging applications generate massive amounts of textual data every day. Analyzing this data can provide valuable insights into user behavior, opinions, and trends. However, raw text often contains noise such as punctuation, stop words, and inconsistent formatting, which makes preprocessing an essential step before applying any Natural Language Processing techniques.

In addition to communication data, textual information from news articles, blogs, and customer reviews plays a crucial role in decision-making processes. Businesses rely on this data to understand customer sentiment, improve products, and enhance user experience. Techniques like tokenization, stemming, lemmatization, and word embeddings help convert unstructured text into meaningful representations. By applying these methods, machines can identify patterns, extract key information, and support tasks such as classification, clustering, and recommendation systems.
"""]

bow = CountVectorizer()
bow_matrix = bow.fit_transform(text)

print("Vocabulary:\n", bow.get_feature_names_out())
print("\nBoW Matrix:\n", bow_matrix.toarray())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(text)

print("Vocabulary:\n", tfidf.get_feature_names_out())
print("\nTF-IDF Matrix:\n", tfidf_matrix.toarray())
```

```python
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
```

```python
text2 = """Artificial Intelligence and Machine Learning are rapidly growing technologies in today's world. Many companies are investing heavily in AI research and development. AI helps in automation, data analysis, prediction, and decision making. Machine learning is a subset of AI that focuses on learning from data without being explicitly programmed.

Deep learning is a specialized branch of machine learning that uses neural networks with multiple layers. These models are capable of understanding complex patterns in data such as images, speech, and text. Natural Language Processing allows machines to understand human language and interact with users in a meaningful way.

In industries like healthcare, AI is used for disease prediction and diagnosis. In finance, it helps detect fraud and automate trading. In transportation, self-driving cars are being developed using AI. Education systems are also integrating AI to provide personalized learning experiences for students.

However, there are concerns about job displacement due to automation. Experts believe that AI will create new job opportunities while replacing repetitive tasks. Ethical considerations, data privacy, and bias in AI systems are also important challenges that need to be addressed carefully."""
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
tokens = word_tokenize(text2.lower())
sentences = [tokens]
```

```python
model = Word2Vec(sentences, vector_size=100, window=5, min_count=1)
```

```python
print("ai vs machine:", model.wv.similarity('ai', 'machine'))
print("data vs analysis:", model.wv.similarity('data', 'analysis'))
print("learning vs deep:", model.wv.similarity('learning', 'deep'))
```

```python
print("Similar to ai:\n", model.wv.most_similar('ai'))
print("Similar to data:\n", model.wv.most_similar('data'))
```
